```python
import pandas as pd
import numpy as np

def clean_price(value):
    if pd.isna(value):
        return np.nan

    value = str(value).lower().strip()

    #remove currency symbols and text
    value = value.replace('₹','').replace('inr','').strip()

    #handle 'k' (thousands)
    if 'k' in value:
        return float(value.replace('k',''))*1000

    #handle normal numbers
    try:
        return float(value)
    except:
        return np.nan
```

```python
df["price_cleaned"] = df["price"].apply(clean_price)
```

# transport cost column

```python
df["transport_cost"].unique()
```

```python
import numpy as np

def clean_transport(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x).lower().strip()
    
    # remove text
    x = x.replace("rs", "").replace("₹", "")
    
    if x == "unknown":
        return np.nan
    
    try:
        return float(x)
    except:
        return np.nan

df["transport_cost_clean"] = df["transport_cost"].apply(clean_transport)
```

```python
df["transport_cost_clean"].dtype
df["transport_cost_clean"].isna().sum()
df[["transport_cost", "transport_cost_clean"]].head(20)
```

# Clean discount column

```python
def clean_discount(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x).lower().strip()
    
    # text to number
    if x == "ten":
        return 10
    
    try:
        return float(x)
    except:
        return np.nan

df["discount_clean"] = df["discount"].apply(clean_discount)
```

```python
# Filter the dataframe to only look at rows where the discount is missing
missing_discounts = df_clean[df_clean["discount"].isna()]

# Check the distribution across suppliers
print("Missing Discounts by Supplier:")
print(missing_discounts["supplier"].value_counts(dropna=False))

print("\n" + "-"*30 + "\n")

# Check the distribution across cities
print("Missing Discounts by City:")
print(missing_discounts["city"].value_counts(dropna=False))
```

# converting the discounts nan to 0 since the nan spread evenly and there is no single vendor or city with higest nan

```python
# Fill missing discounts with 0
df_clean["discount"] = df_clean["discount"].fillna(0)

# Quick check to confirm they are gone
print("Remaining missing discounts:", df_clean["discount"].isna().sum())
```

# Rating column checking between 1 and 5

```python
# Check the unique values and their counts to ensure there are no weird ratings (like 6.0 or -1.0)
print(df_clean["customer_rating"].value_counts(dropna=False).sort_index())
```

# Removing nan from both supplier and warehouse replace with unknown

```python
# Standardize missing values in warehouse
df_clean["warehouse"] = df_clean["warehouse"].fillna("Unknown")

# Standardize missing values in supplier (filling NaNs and capitalizing existing "unknown" strings)
df_clean["supplier"] = df_clean["supplier"].fillna("Unknown")
df_clean["supplier"] = df_clean["supplier"].replace("unknown", "Unknown")

# Quick check to confirm
print("Missing warehouses:", df_clean["warehouse"].isna().sum())
print("Missing suppliers:", df_clean["supplier"].isna().sum())
```

```python
df_clean["quantity"].isna().sum()
```

# Transport cost column

```python
# Filling Transport cost with grouped median since every city has its median transportation cost 
```

```python
# Check the median transport cost for each city
print("Median Transport Cost by City:")
print(df_clean.groupby("city")["transport_cost"].median())

print("\n" + "-"*30 + "\n")

# Let's also look at the overall median just in case
print("Overall Median Transport Cost:", df_clean["transport_cost"].median())
```

```python
# fill the data with grouped median

# Fill missing transport costs with the median of their respective city
df_clean["transport_cost"] = df_clean["transport_cost"].fillna(
    df_clean.groupby("city")["transport_cost"].transform("median")
)

# Quick check to confirm they are gone
print("Remaining missing transport costs:", df_clean["transport_cost"].isna().sum())
```

```python
df_clean.head()
```

```python
df_clean["delivery_status"].isna().sum()
```

# filling delivery status nan to unknown and also notes nan to No Notes

```python
# 1. Clean 'delivery_status': Make capitalization uniform (Title Case) and handle missing data
df_clean["delivery_status"] = df_clean["delivery_status"].str.strip().str.title()
df_clean["delivery_status"] = df_clean["delivery_status"].fillna("Unknown")

# 2. Clean 'notes': Fill missing comments with "No Notes"
df_clean["notes"] = df_clean["notes"].fillna("No Notes")

# 3. The Final Verification!
print(df_clean.isna().sum())
```
